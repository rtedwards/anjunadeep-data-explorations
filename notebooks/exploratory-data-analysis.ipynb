```r
library(IRdisplay)
library(ggplot2)
library(dplyr)
library(tidyr)
library(knitr)
#library(GGally)
library(broom)
```

```r
tracks.df <- read.csv(file="data/anjunadeep-explorations-track-audio-features.csv", header=TRUE)
```

```r
max_danceability <- tracks.df %>%
    arrange(desc(danceability)) %>%
    head(5)
```

```r
head(max_danceability)
```

```r
max_features <- tracks.df %>%
select(popularity, acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, valence) %>%
    summarise_all(funs(max(., na.rm=TRUE)))

min_features <- tracks.df %>%
select(popularity, acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, valence) %>%
    summarise_all(funs(min(., na.rm=TRUE)))

head(max_features)
```

```r
max_danceability <- tracks.df %>%
    filter(danceability == max_features[,'danceability'])
max_danceability[,'track_id']

track_id <- max_danceability[,'track_id']
track_id

html <- paste('<iframe src="https://open.spotify.com/embed/track/', track_id, '" width="300" height="80" frameborder="0" allowtransparency="true" allow="encrypted-media" align="center"></iframe>', sep="")
display_html(html)
```
